```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import datetime
import keras

from sklearn.utils import shuffle
```

```python
train_path = r'asl_dataset\asl_alphabet_train\asl_alphabet_train'
train_path = os.path.join(os.getcwd(), train_path)
```

```python
#chcecking out the data!
categories = os.listdir(train_path)
for category in categories:
    path = os.path.join(train_path, category)
    for image_path in os.listdir(path):
        print(os.path.join(path, image_path))
        img = cv2.imread(os.path.join(path, image_path), cv2.IMREAD_GRAYSCALE)
        break
    break
```

```python
print(img.shape)
img = cv2.resize(img, (100,100))
plt.imshow(img, cmap = 'gray')
plt.show()
```

```python
#generating the dataset!
x_data = []
y_data = []
categories = os.listdir(train_path)
y = 0
for category in categories:
    path = os.path.join(train_path, category)
    for image_path in os.listdir(path):
        #print(os.path.join(path, image_path))
        img = cv2.imread(os.path.join(path, image_path), cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (100,100))
        x_data.append(img)
        y_data.append(y)
    y += 1
    print(category, "\t done!")
```

```python
x_data_np = np.array(x_data)
y_data_np = np.array(y_data)
x_shuffle, y_shuffle = shuffle(x_data_np, y_data_np, random_state = 0)

print(x_shuffle.shape)
print(y_shuffle.shape)

npx_path = os.path.join(os.getcwd(), r"npx_dataset")
np.save(os.path.join(npx_path, "x_data"), x_shuffle)
np.save(os.path.join(npx_path, "y_data"), y_shuffle)
```
