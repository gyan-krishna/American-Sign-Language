```python
from numpy import expand_dims
from keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
from matplotlib import pyplot
```

```python
img = load_img(r'A.jpg')
data = img_to_array(img)
print(type(data[0][0][0]))
samples = expand_dims(data, 0)
```

```python
pyplot.imshow(img)
```

# Width shift augmentation

```python
datagen = ImageDataGenerator(width_shift_range=[-50,50])

#creating an iterator for the data generator
#will return one batch of augmented images per iteration
it = datagen.flow(samples,       #original image that will be used to create augmented data
                  batch_size=1)  #
```

```python
n = 3
fig, plot = pyplot.subplots(n, n, figsize=(8,8))
for i in range(n):
    for j in range(n):
        batch = it.next()
        image = batch[0].astype('uint8')
        plot[i][j].axes.xaxis.set_ticklabels([])
        plot[i][j].axes.yaxis.set_ticklabels([])
        plot[i][j].imshow(image) 
pyplot.show()
```
